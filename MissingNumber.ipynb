268.Given an array nums containing n distinct numbers in the range [0, n], return the only number in the range that is missing from the array.


```python
def fun(nums : list[int])->int:
    nums.sort()
    for i,v in enumerate(nums):
        if i!=v:
            return v-1

        if v==len(nums)-1:
            return v+1    
```

```python
#using xor operation
def fun(nums : [int])->int:
    n=len(nums)
    #a=[i for i in range(n)]
    x=0
    num_xor=0
    for i in range(n+1):
        x ^=i
    for num in nums:
        num_xor ^=num
    return x ^num_xor
if __name__=="__main__":
    ans=fun([1,3,0])
    print(ans)
```

```python
# we sum array of n elem which is len(nums ) and then subtract it from sum(nums)
def fun(nums : [int])->int:
    n=len(nums)
    sum_all=0
    sum_nums=0
    for i in range(n+1):
        sum_all+=i
    for num in nums:
        sum_nums+=num
    return sum_all-sum_nums
if __name__=="__main__":
    ans=fun([1,3,0])
    print(ans)   
```
